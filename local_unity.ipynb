```
%%writefile build.sh

#!/usr/bin/env bash

set -e
set -x

echo "Building for StandaloneLinux64"

export BUILD_PATH=_BUILD_FOLDER_
mkdir -p $BUILD_PATH

${UNITY_EXECUTABLE:-xvfb-run --auto-servernum --server-args='-screen 0 640x480x24' _UNITY_EXEC_ } \
  -projectPath _PROJECT_PATH_ \
  -quit \
  -nographics \
  -batchmode \
  -buildTarget StandaloneLinux64 \
  -customBuildTarget StandaloneLinux64 \
  -customBuildName _BUILD_NAME_ \
  -customBuildPath $BUILD_PATH \
  -executeMethod BuildCommand.PerformBuild \
  -logFile /dev/stdout

UNITY_EXIT_CODE=$?

if [ $UNITY_EXIT_CODE -eq 0 ]; then
  echo "Run succeeded, no failures occurred";
elif [ $UNITY_EXIT_CODE -eq 2 ]; then
  echo "Run succeeded, some tests failed";
elif [ $UNITY_EXIT_CODE -eq 3 ]; then
  echo "Run failure (other failure)";
else
  echo "Unexpected exit code $UNITY_EXIT_CODE";
fi

ls -la $BUILD_PATH
[ -n "$(ls -A $BUILD_PATH)" ] # fail job if build folder is empty
```

```
#@title Build Pathes
import os

PROJECT_PATH = '/content/svo-games/Project/'  #@param {type:"string"}
BUILD_FOLDER = '/content/svo-games/builds/'  #@param {type:"string"}
BUILD_NAME = 'harvest.x86_64'  #@param {type:"string"}

ENV_BINARY = os.path.join(BUILD_FOLDER, BUILD_NAME)
 

import fileinput

with fileinput.FileInput('build.sh', inplace=True, backup='.bak') as file:
  for line in file:
    line = line.replace('_UNITY_EXEC_', UNITY_EXEC)
    line = line.replace('_PROJECT_PATH_', PROJECT_PATH)
    line = line.replace('_BUILD_NAME_', BUILD_NAME)
    line = line.replace('_BUILD_FOLDER_', BUILD_FOLDER)
    print(line, end='')
```
